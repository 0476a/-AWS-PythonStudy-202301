```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd_tips = sns.load_dataset('tips')
print(pd_tips)
```

```python
# 똑같은데 좀 풀어놓은 느낌임!
sns.catplot(data=pd_tips, x='day', y='total_bill', jitter = 0.2, hue='smoker', dodge=True)
# sns.relplot(data=pd_tips, x='day', y='total_bill')
```

```python
sns.catplot(data=pd_tips, x='day', y='total_bill', hue='smoker', dodge=True, kind='swarm')
```

```python
sns.boxplot(data=pd_tips, x='day', y='total_bill', hue='smoker')
```

```python
# sns.catplot(data=pd_tips, x='total_bill', y='day', hue='sex', kind='violin')
# sns.catplot(data=pd_tips, x='total_bill', y='day', hue='sex', kind='violin', bw=.15, cut=0)
# sns.catplot(data=pd_tips, x='day', y='total_bill', hue='sex', kind='violin', split=True)
sns.catplot(data=pd_tips, x='day', y='total_bill', hue='sex', kind='violin',inner='stick', split=True)
```

```python
# sns.catplot(data=pd_tips, x='day', y='total_bill', hue='sex', kind='bar')
sns.catplot(data=pd_tips, x='day', y='total_bill', hue='sex', kind='point')
```
