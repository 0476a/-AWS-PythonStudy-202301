## 인구당 사설학원 수
### 시각화 전략
* 시각화 목표 : 각 구당 인구와 사설학원의 관계
* 시각화의 범위 : 부산광역시내에 2010~2022년까지의 주민등록인구, 사설학원수, 인구 천명당 사설학원수 데이터를 사용할 것이다.
* 시각화의 개발 절차 : kosis.kr에서 "인구 천명당 사설학원수" 데이터 확보, 분석 후 시각화
### 시각화 요구사항
* 사용자 수준 / 취향 / 목적 : 부산내에 학원을 차리고 싶어하는 예비 원장님들이 어느 지역에 학원을 개업하면 성공할 수 있을지 미리 파악해볼 수 있습니다. 또한 학원 취업을 원하는 학원 강사님들이 어느 지역에서 학원에서 취업하기 쉽고 학생들의 유입이 많은 학원에 취업할 수 있을지 미리 파악해볼 수 있습니다. 마지막으로 학원을 다니고 싶어하는 수강생들이 학원을 선택하는데 있어서 도움이 될 수 있는 지표입니다.
* 시나리오 / 스토리보드 개발 : 과거에 비해 사교육의 비중이 매우 증가 한 것으로 보이는 추세입니다. 사교육 학원의 수가 지금까지도 증가하고 있는 추세인지 파악해보고 만약에 올랐다면 모든 지역이 올랐는지 내려간 지역이 있는지 파악해 볼 것입니다. 또한 인구와 사설학원수의 상관도를 구해서 인구가 사설학원수에 얼마나 영향을 미치는지 파악해 볼 것입니다. 마지막으로 인구 천명 당 사설학원수가 각 행정지역별 마다 얼마나 변하였는지도 알아볼 것 입니다.

```python
# 필요 함수 선언 및 한글패치
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list
```

```python
pd_data = pd.read_csv('인구_천명당_사설학원수_시도_시_군_구__20230313164400.csv', encoding='cp949')
pd_data.columns = ['행정구역별', '시점', '인구 천명당 사설학원수', '사설학원수', '주민등록인구']
```

```python
day_data = pd_data.groupby('시점').sum()['사설학원수']
```

```python
day_data
day_data = pd.DataFrame(day_data)
day_data = day_data.reset_index()
```

## 부산광역시 년도별 사설학원 수 변화

```python
# 부산광역시 년도별 사설학원 수 변화
line1 = sns.lineplot(data=day_data, x = '시점', y = '사설학원수')
line1.set_title('부산광역시 년도별 사설학원 수 변화')
```

#### 부산광역시에 사설학원수는 최근에 들어갈 수록 증가하는 것을 확인할 수 있다.

## 각 행정구역별 사설학원 수 변화 추이

```python
line2 = sns.lineplot(data=pd_data, x = '시점', y = '사설학원수', hue='행정구역별')
line2.set_title('행정구역별 사설 학원수 변화')
# 행정구역이 너무 많으므로 구분하기에 어려움이 있다.
```

#### 각각의 행정구역의 학원수가 증가하는 곳이 있고 반대로 감소하는 것이 있는 것을 확인할 수 있다.
#### 다만 행정구역의 수가 많아서 시각적으로 확인하는데 어려움이 있다.

## Top 7의 행정구역 학원수 변화 추이

```python
print(sorted(pd_data.groupby('행정구역별').mean()['사설학원수']))
print(pd_data.groupby('행정구역별').mean())
# top5개만 체크를 해보면 해운대구, 부산진구, 동래구, 북구, 금정구, 남구, 수영구 이다.
```

```python
filter1 = (pd_data['행정구역별'] == '해운대구') | (pd_data['행정구역별'] == '부산진구') | (pd_data['행정구역별'] == '동래구') | (pd_data['행정구역별'] == '북구') | (pd_data['행정구역별'] == '금정구') | (pd_data['행정구역별'] == '남구') | (pd_data['행정구역별'] == '수영구')
top7_pd_data = pd_data[filter1]
# print(top5_pd_data)
line3 = sns.lineplot(data=top7_pd_data, x = '시점', y = '사설학원수', hue='행정구역별')
line3.set_title('top7 행정구역별 변화 추이')
```

#### Top5의 행정구역을 추려서 변화추이를 알아볼 수 있다.
#### 해운대구는 꾸준히 증가하고 있고 사설학원수가 가장 많다.
#### 부산진구는 과거에는 많았지만 현재는 줄어들고 있는 추세이다.
#### 동래구는 꾸준히 증가하면서 현재는 부산진구를 따라 잡은 것을 확인할 수 있다.
#### 북구는 꾸준히 감소하고 있는 추세이다.
#### 금정구와 남구는 변화가 미미한 것을 확인할 수 있다.
#### 수영구는 가장 낮은 값을 나타내며 점점 줄어들고 있는 것을 확인할 수 있다.

## 반대로 가장 낮은 Top7 학원 수 변화 추이

```python
print(sorted(pd_data.groupby('행정구역별').mean()['사설학원수']))
print(pd_data.groupby('행정구역별').mean())
# 반대 top5개만 체크를 해보면 중구, 동구, 영도구, 서구, 강서구, 기장군, 사상구 이다.
```

#### 강서구와 기장군을 보면 처음에는 학원수가 작았는데 계속 올라가는 추세가 보인다.
* 해당 지역에 학원이 많이 생기는 추세라는 것을 파악할 수 있다.
