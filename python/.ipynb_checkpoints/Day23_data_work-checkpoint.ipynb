### 데이터산업의 기술등급별 데이터직무 인력 현황

* 데이터 직무별 axe 생성

* 하나의 figure에, 총 8개 axe가 바둑판 형식으로 배치되는 형태. (4 col x 2 rows), figure 사이즈는 각자 알아서.

```python
pd_data = pd.read_csv('데이터산업의_기술등급별_데이터직무_인력_현황_20230303160526.csv', encoding = 'cp949')


pd_data.drop('기술등급별(1)', axis = 1, inplace=True)
pd_data.columns = ['데이터직무별', '시점', '데이터 처리', '데이터 구축']
pd_data
```

```python
fig, axes = plt.subplots(2,4, figsize = (24,10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)
fig.suptitle('데이터산업의 기술등급별 데이터직무 인력 현황', fontsize = 20)


jobs = ['데이터 아키텍트', '데이터 개발자', '데이터 엔지니어', '데이터 분석가', '데이터베이스 관리자', '데이터 과학자', '데이터 컨설턴트', '데이터 기획자']
for i, job in enumerate(jobs):
    row = i // 4
    col = i % 4
    df = pd_data[pd_data['데이터직무별'] == job]
    axes[row, col].plot(df['시점'], df[['데이터 처리', '데이터 구축']])
    axes[row, col].set_title('{}_raw'.format(job), fontsize = 20)
    axes[row, col].set_xticks(df['시점'])
    axes[row, col].set_yticks(np.linspace(0,25000,6))
```
