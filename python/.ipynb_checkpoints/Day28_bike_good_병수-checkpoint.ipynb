### 부산광역시 이륜차 신고대수
* 출처 : 국토교통부, 자동차등록현황보고, 「이륜차신고현황_시도별 」

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list
```

```python
pd_data = pd.read_csv('이륜오토바이.csv')
pd_data.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
pd_data
```

```python
pd_data.dtypes
```

```python
df1 = pd_data.pivot('지역', '소유', '소계')
```

```python
h1 = sns.heatmap(df1, annot=True, fmt='.0f')
h1.set_title('지역별 소유현황 히트맵')
```

```python
bar1 = sns.barplot(data = pd_data, x='지역', y='소계', hue='소유')
bar1.figure.set_size_inches(6.4*2,4.8)
bar1.set_title('지역별 소유한 오토바이 계수')
```

```python
filter1 = pd_data['소유'] == '관용'
index = pd_data[filter1].index
# print(pd_data[filter1])
# print(pd_data['지역'])
fig, axe = plt.subplots(figsize=(6.4*3, 4.8))
axe.pie(pd_data.iloc[index, 7], labels=pd_data['지역'].unique(), autopct='%.1f')
axe.set_title('부산 각 구별 오토바이 비율')
```

```python
pd_data2 = pd_data.set_index(['년','월','시점','지역','소유']).unstack(level=4).stack(level=0).reset_index()
filter1 = pd_data2['level_4'] == '소계'
print(pd_data2[filter1])
scatter1 = sns.scatterplot(data=pd_data2[filter1], x='관용', y='자가용', hue = '지역')
scatter1.set_title('부산 구역별 관용과 자가용의 상관도')
```

```python
si_data = pd.read_csv('이륜차신고현황_시도별_20230310154713(5대광역시).csv', encoding='cp949', header=1)
si_data.drop('시군구(1)',axis=1, inplace=True)
si_data
```

```python
si_data.insert(0, '년', si_data['시점'].apply(lambda x: str(x).split('.')[0]))
si_data.insert(1, '월', si_data['시점'].apply(lambda x: str(x).split('.')[1]))
```

```python
si_data['시도명(1)'].unique()
```

```python
fig = plt.figure(layout='tight', figsize=(6.4*2, 4.8*3))
ax1 = fig.add_subplot(4,1,1)
ax1.set_title('년도별 서울 오토바이 소계 현황')
ax2 = fig.add_subplot(4,2,3)
ax2.set_title('년도별 부산 오토바이 소계 현황')
ax3 = fig.add_subplot(4,2,4)
ax3.set_title('년도별 대구 오토바이 소계 현황')
ax4 = fig.add_subplot(4,2,5)
ax4.set_title('년도별 인천 오토바이 소계 현황')
ax5 = fig.add_subplot(4,2,6)
ax5.set_title('년도별 광주 오토바이 소계 현황')
ax6 = fig.add_subplot(4,2,7)
ax6.set_title('년도별 대전 오토바이 소계 현황')
ax7 = fig.add_subplot(4,2,8)
ax7.set_title('년도별 울산 오토바이 소계 현황')

Seoul = si_data['시도명(1)']=='서울'
Busan = si_data['시도명(1)']=='부산'
Daegu = si_data['시도명(1)']=='대구'
Incheon = si_data['시도명(1)']=='인천'
Gwangju = si_data['시도명(1)']=='광주'
Daejeon = si_data['시도명(1)']=='대전'
Ulsan = si_data['시도명(1)']=='울산'
sns.pointplot(data=si_data[Seoul], x = '년', y= '소계' , ax=ax1)
sns.pointplot(data=si_data[Busan], x = '년', y= '소계' , ax=ax2)
sns.pointplot(data=si_data[Daegu], x = '년', y= '소계' , ax=ax3)
sns.pointplot(data=si_data[Incheon], x = '년', y= '소계' , ax=ax4)
sns.pointplot(data=si_data[Gwangju], x = '년', y= '소계' , ax=ax5)
sns.pointplot(data=si_data[Daejeon], x = '년', y= '소계' , ax=ax6)
sns.pointplot(data=si_data[Ulsan], x = '년', y= '소계' , ax=ax7)
```
