```python
import numpy as np
import math

x1 = 3
y1 = 10

x2 = 5
y2 = 25

a = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print(a)    # 15.132745950421556

# def dist(p1, p2):      # python list version
#     return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

def dist(p1, p2):      # python list version
    print('x:', (p2[0]-p1[0]), (p2[0]-p1[0])**2)
    print('y:', (p2[1]-p1[1]), (p2[1]-p1[1])**2)
    return math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)

def dist_np(p1, p2):    # [3,10]   [5,25]
    i = p2-p1    # [2,15]
    i = sum(i*i)  # [4, 225] 
    return math.sqrt(i) # 229 루트
    
    
def dist_np2(p1, p2): 
    return math.sqrt(sum((p2-p1)**2))

b = dist([x1, y1], [x2, y2])
print(b)     # 15.132745950421556

c = dist_np(np.array([x1, y1]), np.array([x2, y2]))
print(c)     # 15.132745950421556
```

```python
def my_split(s):
    block_start = False
    start_index = 0
    ret_list=[]
    for i, c in enumerate(s):
        if block_start==False:
            if c==',':
                ret_list.append(s[start_index:i])
                start_index=i+1
            elif c=='"':
                block_start=True
                start_index = i
        else:
            if c=='"':
                block_start=False
    if s[-1]!=',':
        ret_list.append(s[start_index:])
    return ret_list
```

```python
csv_data = []
with open('전국자동차정비업체표준데이터.csv') as f:
    for line in f:
        csv_data.append(my_split(line))
print(csv_data[:3])

# 인덱스 크기 확인
len_list=[]
for e in csv_data:
    len_list.append(len(e))
print(set(len_list))
```
