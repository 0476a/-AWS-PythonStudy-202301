```python
list_a = [i for i in range(10)]
list_b = [i for i in range(100, 110)]

print(list_a)
print(list_b)

# in
if 8 in list_a:
    print('True')
else:
    print('False')

# not in
if 10 not in list_a:
    print('True')
else:
    print('False')

    
# concat
list_c = list_a + list_b
print(list_c)

# * # 횟수만큼 반복되서 나옴!
list_a * 2     # 똑같음  2 * list_a



```

```python
# index 
list_d = list('abcdefghicjklmn')
print(list_d)
print(list_d[3])
print(list_d[3 : 6]) # start : stop # stop 한칸 앞에 까지
print(list_d[3 : 10 : 2]) # start : stop : step # step 만큼 이동
# ":" 사용

# len, min, max
print(len(list_d))
print(min(list_d))
print(max(list_d))

# index
c_index = list_d.index('c')
print(c_index)
list_d[c_index] = 'x'
print(list_d)

c_index = list_d.index('c', 3, 10) # 인데스 3에서 시작해서 10에서 멈춰라 (3 ~ 9) 그중 C를 찾아라
print(c_index)

#count
print('count: ', list_d.count('c'))
```

```python
# 내가한 거

s = 'While The Python Language Reference describes the exact syntax and semantics of the Python language, this library reference manual describes the standard library that is distributed with Python. It also describes some of the optional components that are commonly included in Python distributions.'
print(s)
list_s = list(s)
print(list_s)

# len, min, max
print(len(list_s))
print(min(list_s))
print(max(list_s))

set1 = set(list_s)
print(set1)

# index
c_index = list_s.index('c')
print(c_index)

# count
print('"c" count: ', list_s.count('c'))
print('"W" count: ', list_s.count('W'))
print('공백count: ', list_s.count(' '))
print('"." count: ', list_s.count('.'))
```

```python
# 풀이
s = 'While The Python Language Reference describes the exact syntax and semantics of the Python language, this library reference manual describes the standard library that is distributed with Python. It also describes some of the optional components that are commonly included in Python distributions.'
print(s.lower())
list_s = list(s.lower())
# print(list_s)
set_s = set(list_s)
print(set_s)
exclude_list = ['.', ' ', ',']
for i in set_s:
    if i not in exclude_list:
        print(i, list_s.count(i))
```

```python
list_a = [1, 2, 3]
list_b = list('abc')
d1 = {'a':1, 'b':2, 'c':3}
print(list_a)
print(list_b)
print(d1)

# 리스트 합쳐서 dict 만들기
d2 = dict(zip(list_b, list_a))
print(d2)
print(d2.items())

d3 = dict([('a', 1), ('b', 2), ('c', 3)])
print(d3)

# 키값으로 벨류값 조회
print(d3['a'])
print(d3['c'])
print(d3.get('c'))
# print(d3['x'])
# 값이 없으면 None이 출력
print(d3.get('x'))

# 새로운 값 할당
d3['y'] = 15
print(d3)
```

```python
for i in d3:
    print(i, d3[i])
    
print(d3.keys())
print(d3.values())

for i in d3.values():
    print(i)
    
print(d3)
d3['y'] = 20
print(d3)
```
