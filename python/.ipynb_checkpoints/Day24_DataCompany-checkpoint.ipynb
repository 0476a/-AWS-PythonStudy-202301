### 기획
* 어떤 내용(메시지)을 시각화 할 것인가?
    - 데이터 산업별 시장규모 성장 추이   ---> line (ax.plot)
    - 2015, 2021년도 데이터 산업별 구성비 비교   ---> pie
### 설계
* 해당 메시지를 시각화 하기 위해 어떠한 플롯을 어떻게 위치시킬 것인가?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']

font_path = "C:\Windows\Fonts\gulim.ttc"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list

pd_data = pd.read_csv('데이터산업_시장규모_20230306153053.csv', encoding = 'cp949')
print(pd_data)
```

```python
pd_data.loc[6, '시점'] = '2021'
print(pd_data)
```

```python
fig = plt.figure(figsize=(15,7))
a = fig.add_subplot(2,1,1)
b = fig.add_subplot(2,2,3)
c = fig.add_subplot(2,2,4)

col_names = pd_data.columns[-3:]
xs = pd_data['시점']
ys = pd_data[col_names]

a.plot(xs, ys.iloc[:, 0], 'rv--')
a.plot(xs, ys.iloc[:, 1], 'bo--')
a.plot(xs, ys.iloc[:, 2], 'g^--')
a.set_yticks(np.linspace(0,100000, 3))
a.set_title('년도별 데이터산업 시장규모 증가 추이')
a.legend(col_names)

b.pie(pd_data.iloc[0, 1:], labels=col_names, autopct='%1.1f%%')
b.set_title('2015년 구성비')

c.pie(pd_data.iloc[6, 1:],  labels=col_names, autopct='%1.1f%%')
c.set_title('2021년 구성비')
```
