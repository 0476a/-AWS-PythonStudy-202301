```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd_rawdata = pd.read_csv('전국소방용수시설표준데이터.csv', encoding='cp949')

print(pd_rawdata.columns)
column_selection = ['시도명', '소재지도로명주소', '위도', '경도']
hydrant_data = pd_rawdata[column_selection]
print(hydrant_data.head())
```

```python
pd_rawdata = pd.read_csv('전국소방자동차전용구역표준데이터.csv', encoding='cp949')

# print(pd_rawdata.columns)
column_selection = ['시도명', '소재지도로명주소', '위도', '경도']
parking_data = pd_rawdata[column_selection]
print(parking_data.head())
```

```python
df1 = pd.concat((hydrant_data, parking_data), axis = 0, keys=['hydrant', 'parking'])
df1.reset_index(inplace=True)
df2 = df1.drop(['level_1'], axis = 1)
df2.columns = ['종류', '시도명', '소재지도로명주소', '위도', '경도']
df2
```

```python
g1 = df2[df2['종류'] == 'hydrant'].groupby('시도명')
s1 = g1['종류'].count()
s1.name = 'hydrant'
print(s1)

g2 = df2[df2['종류'] == 'parking'].groupby('시도명')
s2 = g2['종류'].count()
s2.name = 'parking'
print(s2)
```
