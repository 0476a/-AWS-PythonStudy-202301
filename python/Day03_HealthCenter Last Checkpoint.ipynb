```python
csv_data = []
with open('전국건강증진센터표준데이터.csv', encoding = 'cp949') as f:
    for line in f:
        csv_data.append(line.split(','))
print(csv_data[:3])
```

```python
name_list = []
for e in csv_data[1:]:
    if len(e[2]) > 0:
        name_list.append(e[2].split()[0])
        #print(e[2].split()[0])
    #print(e[2].split()[0])
    #print(e[2].split())
print(name_list)
```

```python
for n in name_list:
    if n[0] == '"':
        print(n[1:])
    else:
        print(n)
```

```python
for i, n in enumerate(name_list):
    name_list[i] = n.lstrip('"')
    #print(n.lstrip('"'))
print(name_list[:10])
```

```python
# 내가한거!
# 빈도수 dicionary 반환
# 1. raw_list를 셋으로 변환
# 2. 셋을 순환하면서, 각각 몇개씩 raw_list에 잇는지 카운트
# 3. 1,2에서 얻은 key, value 값으로 dictionary 만들어 return.
def unique_count(raw_list, sort = False):
    raw_list1 = []
    count_list = []
    for n in set(raw_list):
        raw_list1.append(n)
        count_list.append(raw_list.count(n))
    raw_dict = dict(zip(raw_list1, count_list))
    
    raw_sorted = dict(sorted(raw_dict.items(), key = lambda x: x[1], reverse = sort))
    
    return raw_sorted
    

name_dict = unique_count(name_list,True)
print(name_dict)
```
