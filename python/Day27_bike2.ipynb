```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

current_font_list = matplotlib.rcParams['font.family']
font_path = "C:\Windows\Fonts\HMKMRHD.TTF"
kfont = matplotlib.font_manager.FontProperties(fname=font_path).get_name()
matplotlib.rcParams['font.family'] = [kfont] + current_font_list
```

```python
pd_data = pd.read_csv('이륜차신고현황_시도별_20230309170733(풀이용).csv', encoding='cp949')
pd_data
```

```python
pd_data.drop(['시점'], axis = 1, inplace=True)
pd_data.drop([0], axis = 0, inplace=True)
```

```python
pd_data
```

```python
pd_data['경형'] = pd_data['경형'].astype('int64')
pd_data['소형'] = pd_data['소형'].astype('int64')
pd_data['중형'] = pd_data['중형'].astype('int64')
pd_data['대형'] = pd_data['대형'].astype('int64')
pd_data['경형.1'] = pd_data['경형.1'].astype('int64')
pd_data['소형.1'] = pd_data['소형.1'].astype('int64')
pd_data['중형.1'] = pd_data['중형.1'].astype('int64')
pd_data['대형.1'] = pd_data['대형.1'].astype('int64')
pd_data['경형.2'] = pd_data['경형.2'].astype('int64')
pd_data['소형.2'] = pd_data['소형.2'].astype('int64')
pd_data['중형.2'] = pd_data['중형.2'].astype('int64')
pd_data['대형.2'] = pd_data['대형.2'].astype('int64')
```

```python
pd_data.dtypes
```

```python
fig = plt.figure(figsize=(6.4*3, 4.8*2))
a = fig.add_subplot(2,2,1)
b = fig.add_subplot(2,2,2)
c = fig.add_subplot(2,2,3)
d = fig.add_subplot(2,2,4)

sns.barplot(data = pd_data, x='시군구(1)', y='경형', ax= a)
sns.barplot(data = pd_data, x='시군구(1)', y='소형', ax= b)
sns.barplot(data = pd_data, x='시군구(1)', y='중형', ax= c)
sns.barplot(data = pd_data, x='시군구(1)', y='대형', ax= d)
```
