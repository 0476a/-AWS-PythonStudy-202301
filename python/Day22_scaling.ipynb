```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

url = 'https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt'
pd_data = pd.read_csv(url, sep = '\t')
print(pd_data.head())
print(pd_data.shape)
```

```python
_, axe = plt.subplots()
axe.plot(pd_data['BMI'])
axe.plot(pd_data['BP'])
```

```python
# normalization    : (x-x_min)/(x_max-x_min)
pd_data['BMI'].describe()
bmi_min = pd_data['BMI'].min()
bmi_max = pd_data['BMI'].max()
pd_data['BMI_norm'] = pd_data['BMI'].apply(lambda x: (x-bmi_min)/(bmi_max-bmi_min))

pd_data['BP'].describe()
bp_min = pd_data['BP'].min()
bp_max = pd_data['BP'].max()
pd_data['BP_norm'] = pd_data['BP'].apply(lambda x: ((x-bp_min)/(bp_max-bp_min)))


_, axe = plt.subplots()
axe.plot(pd_data['BMI_norm'], color = 'green')
axe.plot(pd_data['BP_norm'], color = 'yellow')

print(pd_data['BP_norm'].describe())
```

```python
# standardization: (x-x_mean)/x_std
# -   () .

pd_data['BMI'].describe()
bmi_mean = pd_data['BMI'].mean()
bmi_std = pd_data['BMI'].std()
pd_data['BMI_std'] = pd_data['BMI'].apply(lambda x: (x-bmi_mean)/bmi_std)

pd_data['BP'].describe()
bp_mean = pd_data['BP'].mean()
bp_std = pd_data['BP'].std()
pd_data['BP_std'] = pd_data['BP'].apply(lambda x: (x-bp_mean)/bp_std)

_, axe = plt.subplots()
axe.plot(pd_data['BMI_std'], color = 'red')
axe.plot(pd_data['BP_std'], color = 'black')

print(pd_data['BP_std'].describe())
```
