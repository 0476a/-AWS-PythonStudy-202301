```python
csv_data = []
csv_data_int = []
with open('lott.CSV') as f:
    for line in f:
        csv_data.append(line[:-1].split(','))
        #csv_data.append(line.replace('\n', '').split(','))
print(csv_data[:3])
int(csv_data[0][-1])
```

```python
import math
def mean(num_list):
    return sum(num_list) / len(num_list)

def dev(num_list):    # 편차 : 평균 값 대비 차이 값
    m = mean(num_list)
    return [x-m for x in num_list]

def var(num_list):    # 분산 : 편차에 대표성있는 값
    n = len(num_list)
    d = dev(num_list)
    return sum([x*x for x in d]) / (n-1)

def stdev(num_list):   # 표준편차 : 분산에서 단위가 맞지 않는 것을 바꿔주는 작업
    return math.sqrt(var(num_list))
```

```python
avg_list = []
stdev_list = []
for e in csv_data:
    for i,s  in enumerate(e): 
        e[i] = int(s)
# print(csv_data)


for i in csv_data:
    avg_list.append(mean(i))
    stdev_list.append(stdev(i))
    
# 회차 마다 평균값 
print('회차 마다 평균값')
for e, s in enumerate(avg_list):
    print(e + 1, '회차' , s)

# print(avg_list)
# print(stdev_list)s
        
```

```python
# 회차 마다 표준편차 
print('회차 마다 표준편차')
for e, s in enumerate(stdev_list):
    print(e + 1, '회차', s)
```

```python
# 1. 전체 회차를 통틀어 빈도수 (mode): 숫자별 카운트 (몇번이나 나왔나..) 1 ~ 45
# 2. 회차별 - 평균값, 표준편차

lott_list = []
lott_num_list = []
lott_count_list = []

for i in csv_data:
    for k in i:
        lott_list.append(k)
# print(lott_list)

lott_set = set(lott_list)

for i in lott_set:
    lott_num_list.append(i)
    lott_count_list.append(lott_list.count(i))
max_number = max(lott_count_list)
lott_dict = dict(zip(lott_count_list, lott_num_list))

# 해당 번호들의 빈도 수
print(lott_dict)
print('빈도수가 가장 많은 번호와 횟수 : ' + lott_dict[max_number], max_number)


```
