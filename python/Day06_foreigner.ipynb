```python
csv_data = []
with open('국적별_외국인_현황_20230207151126.csv') as f:
    for line in f:
        csv_data.append(line.split(","))
# print(csv_data[:10])

# 1번 인덱스 삭제
csv_data.remove(csv_data[1])

# 부산 광역시 삭제
for i, frgn in enumerate(csv_data):
    csv_data[i].remove(frgn[1])
print(csv_data[:10])
```

```python
# 2022년 11월 인덱스 구하기
year_list = []
for frgn in csv_data:
    year_list.append(frgn[0])
print(year_list)
year_index = year_list.index('"2022.11"')
print(year_index)
```
